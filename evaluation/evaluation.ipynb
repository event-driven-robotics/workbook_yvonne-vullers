```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas 
import cv2
from scipy import interpolate
import os
```

```python
user = 13
eyenum = 1
versionfc = "eros_full_cut"
versionff = "eros_full_full"
versioncf = "eros_cut_full"
versioncc = "eros_cut_cut"
versionf = "fixed_cut_cut"
save = True

f_errorfc = open("/home/yvullers/Documents/Yvonne/error/error_" + versionfc + ".txt", "a")
f_errorff = open("/home/yvullers/Documents/Yvonne/error/error_" + versionff + ".txt", "a")
f_errorcf = open("/home/yvullers/Documents/Yvonne/error/error_" + versioncf + ".txt", "a")
f_errorcc = open("/home/yvullers/Documents/Yvonne/error/error_" + versioncc + ".txt", "a")
f_errorf = open("/home/yvullers/Documents/Yvonne/error/error_" + versionf + ".txt", "a")

f_timefc = open("/home/yvullers/Documents/Yvonne/frequency/frequency_" + versionfc + ".txt", "a")
f_timeff = open("/home/yvullers/Documents/Yvonne/frequency/frequency_" + versionff + ".txt", "a")
f_timecf = open("/home/yvullers/Documents/Yvonne/frequency/frequency_" + versioncf + ".txt", "a")
f_timecc = open("/home/yvullers/Documents/Yvonne/frequency/frequency_" + versioncc + ".txt", "a")
f_timef = open("/home/yvullers/Documents/Yvonne/frequency/frequency_" + versionf + ".txt", "a")

f_stdfc = open("/home/yvullers/Documents/Yvonne/error/std_" + versionfc + ".txt", "a")
f_stdff = open("/home/yvullers/Documents/Yvonne/error/std_" + versionff + ".txt", "a")
f_stdcf = open("/home/yvullers/Documents/Yvonne/error/std_" + versioncf + ".txt", "a")
f_stdcc = open("/home/yvullers/Documents/Yvonne/error/std_" + versioncc + ".txt", "a")
f_stdf = open("/home/yvullers/Documents/Yvonne/error/std_" + versionf + ".txt", "a")
```

```python
filefc = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/{vers}/center_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum, vers = versionfc)
fileff = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/{vers}/center_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum, vers = versionff)
filecf = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/{vers}/center_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum, vers = versioncf)
filecc = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/{vers}/center_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum, vers = versioncc)
filef = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/{vers}/center_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum, vers = versionf)

ground_truth = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/ground_truth_{usernum}_{eye}.csv".format(usernum = user, eye = eyenum)
roberta = "/home/yvullers/Documents/Yvonne/CenterData/User{usernum}/{eye}/user{usernum}_{eye}.txt".format(usernum = user, eye = eyenum)

df_centerfc = pandas.read_csv(filefc, sep='\s+', header=None)
df_centerff = pandas.read_csv(fileff, sep='\s+', header=None)
df_centercf = pandas.read_csv(filecf, sep='\s+', header=None)
df_centercc = pandas.read_csv(filecc, sep='\s+', header=None)
df_centerf = pandas.read_csv(filef, sep='\s+', header=None)

df_gt = pandas.read_csv(ground_truth, sep='\s+', header=None)
df_roberta = pandas.read_csv(roberta, sep='\s+', header=None)

centerfc = df_centerfc.to_numpy()
centerff = df_centerff.to_numpy()
centercf = df_centercf.to_numpy()
centercc = df_centercc.to_numpy()
centerf = df_centerf.to_numpy()

gt = df_gt.to_numpy()
center_roberta = df_roberta.to_numpy()
```

ERROR CALCULATION

```python
temp_x = (centerfc[:,6])
temp_y = (centerfc[:,7])

f = interpolate.interp1d(gt[:,0], gt, kind = "nearest", axis = 0, fill_value="extrapolate")
gt_int = f(centerfc[:,0])

gt_center_x = (gt_int[:,1]+gt_int[:,3])/2
gt_center_y = (gt_int[:,2]+gt_int[:,4])/2

f_rob = interpolate.interp1d(center_roberta[:,0], center_roberta, kind = "nearest", axis = 0, fill_value="extrapolate")
rob = f_rob(centerfc[:,0])

errorx = abs(gt_center_x - temp_x).mean()
stdx = abs(gt_center_x - temp_x).std()
errory = abs(gt_center_y - temp_y).mean()
stdy = abs(gt_center_y - temp_y).std()

errorx_rob = abs(gt_center_x - rob[:,2]).mean()
errory_rob = abs(gt_center_y - rob[:,1]).mean()
stdx_rob = abs(gt_center_y - rob[:,1]).std()
stdy_rob = abs(gt_center_y - rob[:,1]).std()

string = str(user) + " " + str(eyenum) + " " + str(errorx) + " " + str(errory) + " " + str(errorx_rob) + " " + str(errory_rob) + "\n"
string_std = str(user) + " " + str(eyenum) + " " + str(stdx) + " " + str(stdy) + " " + str(stdx_rob) + " " + str(stdy_rob) + "\n"

if(save == True):
    f_errorfc.write(string)
    f_stdfc.write(string_std)
f_errorfc.close()
f_stdfc.close()
```
